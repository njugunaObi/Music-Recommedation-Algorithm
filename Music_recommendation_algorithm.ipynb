**SPOTIFY MUSIC RECOMMENDATION ALGORITHM**


We are to creating a recommendation algorithm to recommend spotify songs & playlists during the listening process.

The data has been sourced from Kaggle & all the preprocessing steps have been done by me.

Lets get started.


```python
# reading the data

import pandas as pd


data =  pd.read_csv('dataset.csv')

data.head()
```

```python
# going deeper into the info of the data

data.info()

# searching for duplicates

data.duplicated()

# dropping irrelevant columns

data = data.drop(['explicit','danceability','instrumentalness','speechiness'], axis=1)

data.head()
```

Preprocessing the data, we have found that;

  **MODELLING THE DATA**

Time to create the algorithm and model the data.

For this recommendation algorithm i have decided to use a Content based filtering approach. Content-based filtering, uses the features of songs (such as genre, artist, and tempo) to make recommendations. It works by finding songs that are similar to the ones that a user has enjoyed in the past.

I shall be using the genre & album name as the points to train the model upon.

```python
print(data.isnull())
```

**RECOMMENDATION ALGORITHM**

```python
# building the model

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix


# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Replace NaN values with an empty string
data = data.fillna('')

# Construct the TF-IDF matrix as a sparse matrix
tfidf_matrix = csr_matrix(tfidf.fit_transform(data['track_genre']))

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Define a function to get recommendations
def get_recommendations(title, cosine_sim=cosine_sim, df=data):
    # Get the index of the music that matches the title
    idx = data[data['album_name'] == title].index[0]

    # Get the pairwise similarity scores of all music with that music
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the music based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar music
    sim_scores = sim_scores[1:11]

    # Get the music indices
    music_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar music
    return data['album_name'].iloc[music_indices]
```

```python
get_recommendations('Hold On')
```

**MODELLING EXPLANATION & CONCLUSIONS**

1. Started out by checking for any null/NAN values within the dataset. There were none so we went straight into writing the model.

2. First we start by creating a TF-IDF vectorizer to convert the text data into a matrix of TF-IDF features.

3. We then replace any NaN values with an empty string.

4. We now construct the TF-IDF matrix as a sparse matrix to enable the model to run efficiently and save computational power.

5. Then we compute the cosine similarity matrix using the cosine_similarity function from scikit-learn to compute the pairwise cosine similarity scores between all pairs of items in the dataset.

6. Lastly we define a function to get recommendations. This takes as input the album title and returns the top 10 most similar titles based on their cosine similarity scores.

7. Finally we test the function by calling the "get_recommendations" function to see if the algorithm works.
